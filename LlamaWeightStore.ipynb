```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import numpy as np
import os
import json
```

```python
MyHFToken = "" # Put your token, from the hugging face
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id, token=MyHFToken)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    token=MyHFToken
).to("cuda")

model.eval()
```

```python
weigthPath = r""
def SaveTensor(mat, name):
    os.makedirs(weigthPath, exist_ok=True)
    path = weigthPath + "/" + name + ".bin"
    mat = mat.detach().cpu().numpy()
    mat = np.ascontiguousarray(mat)
    mat.tofile(path)
```

```python
def SaveTxtFile(data, filePath):
    with open(filePath, "w") as textFile:
        textFile.write(data)
    print("Save text file")
```

```python
parameterTxt = ""
for name, parmeter in model.named_parameters():
    SaveTensor(parmeter ,name)
    if(parmeter.dim() == 2):
        parameterTxt += name + f":{parmeter.shape[0]},{parmeter.shape[1]}\n"
    else:
        parameterTxt += name + f":{parmeter.shape[0]},1\n"
    print(f"{name} saved")
SaveTxtFile(parameterTxt, os.path.join(weigthPath, "metData.txt"))
```
