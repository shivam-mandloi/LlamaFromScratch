```python
import json
from tqdm import tqdm
```

```python
def ReadJson(filePath):
    with open(filePath, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data

def SaveFile(text, fileName):
    with open(r"C:\Users\shiva\Desktop\IISC\LLAMA\LLAMA Weights\ " + fileName + ".txt", "w", encoding="utf-8") as f:
        f.write(text)
```

```python
filePath = r"C:\Users\shiva\Desktop\IISC\LLAMA\LLAMA Weights\tokenizer.json"
data = ReadJson(filePath)
```

```python
merges = data["model"]["merges"]
vocab = data["model"]["vocab"]
```

```python
fullString = "<^>".join(merges)
SaveFile(fullString, "mergesCpp")
```

```python
txt = " ".join(list(vocab.keys()))
SaveFile(txt, "vocabCPP")
```
